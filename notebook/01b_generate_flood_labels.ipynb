```python
import numpy as np
import matplotlib.pyplot as plt

features = np.load("../data/processed_features.npy")
delta_VH = features[:, :, 1]

print(delta_VH.shape)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(delta_VH, cmap="RdBu")
plt.colorbar(label="VH")
plt.title("Delta VH")
plt.axis("off")
plt.show()
```

```python
threshold = -1.0

flood_labels = np.zeros_like(delta_VH, dtype=np.uint8)
flood_labels[delta_VH < threshold] = 1
```

```python
plt.figure(figsize=(6,6))
plt.imshow(flood_labels, cmap="gray")
plt.title("Automated Flood Labels")
plt.axis("off")
plt.show()
```

```python
np.save("../data/flood_labels.npy", flood_labels)
print("flood_labels.npy saved")
```
