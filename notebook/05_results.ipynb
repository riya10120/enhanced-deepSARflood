```python
import numpy as np
import matplotlib.pyplot as plt

flood_map = np.load("../data/predicted_flood_map.npy")

plt.figure(figsize=(6,6))
plt.imshow(flood_map, cmap="Blues")
plt.title("Final Flood Extent Map")
plt.axis("off")
plt.show()
```

```python
total_pixels = flood_map.size
flood_pixels = np.sum(flood_map == 1)

flood_percentage = (flood_pixels / total_pixels) * 100

print(f"Flooded area percentage: {flood_percentage:.2f}%")
```

```python
construction_suitability = np.zeros_like(flood_map)
construction_suitability[flood_map == 0] = 1  # suitable
```

```python
plt.figure(figsize=(6,6))
plt.imshow(construction_suitability, cmap="Greens")
plt.title("Construction Suitability Map")
plt.axis("off")
plt.show()
```

```python
np.save("../data/construction_suitability.npy", construction_suitability)
print("Final outputs saved")
```
