```python
import ee
ee.Initialize(project="deepsarflood-481610")
print("EE initialized with project")
```

```python
chennai = ee.Geometry.Rectangle([80.15, 12.85, 80.35, 13.15])
```

```python
s1_non = (
    ee.ImageCollection("COPERNICUS/S1_GRD")
    .filterBounds(chennai)
    .filterDate("2023-01-10", "2023-01-20")
    .filter(ee.Filter.eq("instrumentMode", "IW"))
    .filter(ee.Filter.listContains("transmitterReceiverPolarisation", "VH"))
    .select("VH")
    .mean()
)
print("Non-flood image ready")
```

```python
task = ee.batch.Export.image.toDrive(
    image=s1_non,
    description="chennai_non_flood",
    folder="DeepSARFlood",
    fileNamePrefix="non_flood_pic",
    region=chennai,
    scale=10,
    maxPixels=1e13
)

task.start()
print("Export started")
```

```python
collection_mod = (
    ee.ImageCollection("COPERNICUS/S1_GRD")
    .filterBounds(chennai)
    .filterDate("2023-11-15", "2023-12-05")  # wider window
    .filter(ee.Filter.eq("instrumentMode", "IW"))
    .filter(ee.Filter.listContains("transmitterReceiverPolarisation", "VH"))
)

print("Moderate flood image count:",
      collection_mod.size().getInfo())
```

```python
s1_mod = collection_mod.select("VH").mean()
print("Moderate flood image created")
```

```python
task = ee.batch.Export.image.toDrive(
    image=s1_mod,
    description="chennai_moderate_flood",
    folder="DeepSARFlood",
    fileNamePrefix="moderate_flood_pic",
    region=chennai,
    scale=10,
    maxPixels=1e13
)

task.start()
print("Moderate flood export started")
```

```python
collection_sev = (
    ee.ImageCollection("COPERNICUS/S1_GRD")
    .filterBounds(chennai)
    .filterDate("2015-12-01", "2015-12-15")
    .filter(ee.Filter.eq("instrumentMode", "IW"))
    .filter(ee.Filter.listContains("transmitterReceiverPolarisation", "VV"))
)

print("Severe flood VV count:",
      collection_sev.size().getInfo())
```

```python
s1_sev = collection_sev.select("VV").mean()

task = ee.batch.Export.image.toDrive(
    image=s1_sev,
    description="chennai_severe_flood",
    folder="DeepSARFlood",
    fileNamePrefix="severe_flood_pic",
    region=chennai,
    scale=10,
    maxPixels=1e13
)

task.start()
print("Severe flood export started")
```
